#### It seems that we have overfitting here, so we are going to reduse the dimentionality of the data by dropping the columns with high correlation value 

```python
# with the following function we can select highly correlated features
# it will remove the first feature that is correlated with anything other feature

def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

## Important !!!

 - We usually use a threshold with the value (0.85) but the dataset is with high number of features ,
 - So we decided to make it (0.92).
 - After applying the threshold value with 0.92 we decided to reduce it to 0.7 instead, but it might be a risk to do that.
 - So i think we need a domain expert here.

```python

corr_features = correlation(X_train, 0.5)
len(set(corr_features))
```

```python
corr_features
```

```python
X_train_selected = X_train.drop(corr_features,axis=1)
X_test_selected = X_test.drop(corr_features,axis=1)
```
